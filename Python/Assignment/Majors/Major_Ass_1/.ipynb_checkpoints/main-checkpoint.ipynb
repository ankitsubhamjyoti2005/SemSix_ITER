```python
import random
from enum import Enum
```

```python
class Suit(Enum):
    HEARTS = "Hearts"
    DIAMONDS = "Diamonds"
    CLUBS = "Clubs"
    SPADES = "Spades"

class Card:
    def __init__(self, suit: Suit, rank: str):
        self.suit = suit
        self.rank = rank

    def __repr__(self):
        return f"Card({self.suit.value}, {self.rank})"

    def value(self):
        """Returns the Blackjack value of the card"""
        if self.rank in ["Jack", "Queen", "King"]:
            return 10
        elif self.rank == "Ace":
            return 11  # Ace can be 1 or 11, but we handle it dynamically in hand evaluation
        else:
            return int(self.rank)
```

```python
class Deck:
    def __init__(self):
        ranks = ["Ace", "2", "3", "4", "5", "6", "7", "8", "9", "10", "Jack", "Queen", "King"]
        self.cards = [Card(suit, rank) for suit in Suit for rank in ranks]
        self.shuffle()

    def shuffle(self):
        random.shuffle(self.cards)

    def draw_card(self):
        return self.cards.pop() if self.cards else None  # Returns None if the deck is empty

# Test deck creation
deck = Deck()
print(deck.cards[:5])  # Print first five cards to verify shuffling
```

```python
class Hand:
    def __init__(self):
        self.cards = []

    def add_card(self, card):
        self.cards.append(card)

    def total_value(self):
        """Calculates the total hand value, treating Aces as 1 or 11 dynamically"""
        total = sum(card.value() for card in self.cards)
        ace_count = sum(1 for card in self.cards if card.rank == "Ace")

        while total > 21 and ace_count:
            total -= 10  # Convert an Ace from 11 to 1
            ace_count -= 1

        return total

    def __repr__(self):
        return f"Hand: {self.cards} (Total: {self.total_value()})"

# Test hand
hand = Hand()
hand.add_card(Card(Suit.HEARTS, "Ace"))
hand.add_card(Card(Suit.SPADES, "King"))
print(hand)  # Expected output: Ace and King (Total: 21)
```

```python
class Player:
    def __init__(self, name):
        self.name = name
        self.hand = Hand()

    def hit(self, deck):
        card = deck.draw_card()
        if card:
            self.hand.add_card(card)
        return card

    def is_bust(self):
        return self.hand.total_value() > 21

    def __repr__(self):
        return f"{self.name}: {self.hand}"

class Dealer(Player):
    def __init__(self):
        super().__init__("Dealer")

    def should_hit(self):
        return self.hand.total_value() < 17  # Dealer's rule: hit on 16 or less
```

```python
class BlackjackGame:
    def __init__(self, player_names):
        self.deck = Deck()
        self.players = [Player(name) for name in player_names]
        self.dealer = Dealer()
        self.initialize_game()

    def initialize_game(self):
        """Deals two initial cards to each player and the dealer."""
        for _ in range(2):
            for player in self.players:
                player.hit(self.deck)
            self.dealer.hit(self.deck)

    def player_turn(self, player):
        """Handles player actions: hit or stand."""
        while True:
            print(f"\n{player}")
            action = input(f"{player.name}, do you want to 'hit' or 'stand'? ").strip().lower()
            if action == "hit":
                player.hit(self.deck)
                if player.is_bust():
                    print(f"{player.name} busts!")
                    break
            elif action == "stand":
                break

    def dealer_turn(self):
        """Dealer automatically plays based on rules."""
        while self.dealer.should_hit():
            self.dealer.hit(self.deck)

    def determine_winners(self):
        """Determines the winners based on game rules."""
        dealer_total = self.dealer.hand.total_value()
        print(f"\nFinal Dealer Hand: {self.dealer}")

        for player in self.players:
            player_total = player.hand.total_value()
            if player.is_bust():
                print(f"{player.name} loses (busted).")
            elif dealer_total > 21 or player_total > dealer_total:
                print(f"{player.name} wins!")
            elif player_total == dealer_total:
                print(f"{player.name} ties with the dealer.")
            else:
                print(f"{player.name} loses.")

    def play(self):
        """Runs the full game loop."""
        for player in self.players:
            self.player_turn(player)

        self.dealer_turn()
        self.determine_winners()

# Test game with two players
game = BlackjackGame(["Alice", "Bob"])
game.play()
```

```python
def start_multiplayer_blackjack():
    player_count = int(input("Enter the number of players: "))
    player_names = [input(f"Enter name for player {i+1}: ") for i in range(player_count)]

    game = BlackjackGame(player_names)
    game.play()

# Start a multiplayer game
start_multiplayer_blackjack()
```

```python
import random
import pathlib
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from enum import Enum
```

```python
class Suit(Enum):
    HEARTS = "hearts"
    DIAMONDS = "diamonds"
    CLUBS = "clubs"
    SPADES = "spades"

class Card:
    def __init__(self, suit: Suit, rank: str, image_folder: pathlib.Path):
        self.suit = suit
        self.rank = rank.lower()
        self.image_path = image_folder / f"{self.rank}_of_{self.suit.value}.png"

    def __repr__(self):
        return f"{self.rank.capitalize()} of {self.suit.value.capitalize()}"

    def value(self):
        if self.rank in ["jack", "queen", "king"]:
            return 10
        elif self.rank == "ace":
            return 11
        else:
            return int(self.rank)

class Deck:
    def __init__(self, image_folder: pathlib.Path):
        ranks = ["ace", "2", "3", "4", "5", "6", "7", "8", "9", "10", "jack", "queen", "king"]
        self.cards = [Card(suit, rank, image_folder) for suit in Suit for rank in ranks]
        self.shuffle()

    def shuffle(self):
        random.shuffle(self.cards)

    def draw_card(self):
        return self.cards.pop() if self.cards else None

class Hand:
    def __init__(self):
        self.cards = []

    def add_card(self, card):
        self.cards.append(card)

    def total_value(self):
        total = sum(card.value() for card in self.cards)
        ace_count = sum(1 for card in self.cards if card.rank == "ace")

        while total > 21 and ace_count:
            total -= 10  # Convert an Ace from 11 to 1
            ace_count -= 1

        return total

    def is_bust(self):
        return self.total_value() > 21

```

```python
def display_cards(players, dealer):
    num_players = len(players)
    fig, axes = plt.subplots(2, num_players, figsize=(num_players * 2, 5))

    # Ensure axes is always a 2D array (handles case when num_players = 1)
    if num_players == 1:
        axes = [axes]

    # Display player hands in the top row
    for i, player in enumerate(players):
        axes[0, i].set_title(f"{player.name} ({player.hand.total_value()})")
        axes[0, i].axis("off")
        display_hand(axes[0, i], player.hand.cards)

    # Display dealer's hand in the bottom row (centered)
    dealer_ax = fig.add_subplot(2, num_players, num_players + 1)
    dealer_ax.set_title(f"Dealer ({dealer.hand.total_value()})")
    dealer_ax.axis("off")
    display_hand(dealer_ax, dealer.hand.cards)

    plt.tight_layout()
    plt.show()

def display_hand(ax, cards):
    x_offset = 0
    for card in cards:
        if card and card.image_path.exists():
            img = mpimg.imread(card.image_path)
            ax.imshow(img, extent=[x_offset, x_offset + 1, 0, 1])
            x_offset += 1
```

```python
class Player:
    def __init__(self, name):
        self.name = name
        self.hand = Hand()

    def hit(self, deck):
        card = deck.draw_card()
        if card:
            self.hand.add_card(card)
        return card

class Dealer(Player):
    def __init__(self):
        super().__init__("Dealer")

    def should_hit(self):
        return self.hand.total_value() < 17
```

```python
class BlackjackGame:
    def __init__(self, player_names, image_folder):
        self.deck = Deck(image_folder)
        self.players = [Player(name) for name in player_names]
        self.dealer = Dealer()
        self.initialize_game()

    def initialize_game(self):
        """Deals two initial cards to each player and the dealer."""
        for _ in range(2):
            for player in self.players:
                player.hit(self.deck)
            self.dealer.hit(self.deck)

    def player_turn(self, player):
        """Handles player actions: hit or stand."""
        while True:
            display_cards(self.players, self.dealer)
            action = input(f"{player.name}, do you want to 'hit' or 'stand'? ").strip().lower()
            if action == "hit":
                player.hit(self.deck)
                if player.hand.is_bust():
                    print(f"{player.name} busts!")
                    break
            elif action == "stand":
                break

    def dealer_turn(self):
        """Dealer automatically plays based on rules."""
        while self.dealer.should_hit():
            self.dealer.hit(self.deck)

    def determine_winners(self):
        """Determines the winners based on game rules."""
        dealer_total = self.dealer.hand.total_value()
        display_cards(self.players, self.dealer)

        for player in self.players:
            player_total = player.hand.total_value()
            if player.hand.is_bust():
                print(f"{player.name} loses (busted).")
            elif dealer_total > 21 or player_total > dealer_total:
                print(f"{player.name} wins!")
            elif player_total == dealer_total:
                print(f"{player.name} ties with the dealer.")
            else:
                print(f"{player.name} loses.")

    def play(self):
        """Runs the full game loop."""
        for player in self.players:
            self.player_turn(player)

        self.dealer_turn()
        self.determine_winners()
```
